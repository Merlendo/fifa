```python
import json
def data_équipes_dict():
    équipes = {}
    with open("data_équipes.csv", "r", encoding="utf8") as f_data_equipes:
        data_equipes = [line.strip().split(";") for line in f_data_equipes][1:]
    for e in data_equipes:
        équipe = {}
        équipe["gardiens"] = json.loads(e[1])
        équipe["défenseurs"] = json.loads(e[2])
        équipe["milieux"] = json.loads(e[3])
        équipe["attaquants"] = json.loads(e[4])
        équipes[e[0]] = équipe
    
    return équipes
```

```python
def data_joueurs_dict(data_csv):
    with open(data_csv, "r", encoding="utf8") as f_data_joueurs:
        data_joueurs = [line.strip().split(",") for line in f_data_joueurs][1:]

    joueurs = {}
    for j in data_joueurs:
        joueur = {}
        joueur["nationalité"] = j[1]
        joueur["rating"] = j[2]
        joueur["âge"] = j[3]
        joueurs[j[0]] = joueur
        
    return joueurs
```

```python
def joueur_disponibilité(équipes, joueurs):
    joueurs_absents = []
    joueurs_present = []
    for nom in équipes:
        for poste in ["gardiens","défenseurs", "milieux", "attaquants"]:
            for j in équipes[nom][poste]:
                if j not in joueurs:
                    joueurs_absents.append(j)
                else:
                    joueurs_present.append(j)
    return joueurs_absents, joueurs_present
```

```python
def recherche_joueur(joueur):
    import requests
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:108.0) Gecko/20100101 Firefox/108.0',
        'Accept': '*/*',
        'Accept-Language': 'fr,fr-FR;q=0.8,en-US;q=0.5,en;q=0.3',
        # 'Accept-Encoding': 'gzip, deflate, br',
        'Referer': 'https://www.fifaindex.com/',
        'Connection': 'keep-alive',
        'Sec-Fetch-Dest': 'empty',
        'Sec-Fetch-Mode': 'cors',
        'Sec-Fetch-Site': 'same-origin',
        'Sec-GPC': '1',
    }

    params = {
        'query': str(joueur),
    }

    response = requests.get('https://www.fifaindex.com/search/suggest/', params=params, headers=headers)
    j = response.json()
    return j
```

```python
def liens_joueurs(joueurs_absents):
    lien_absent = []
    lien_possible = []
    for i, j in enumerate(joueurs_absents):
        joueur_json = recherche_joueur(j)
        if len(joueur_json) != 0:
            print(f'{i+1}/{len(joueurs_absents)}')
            lien_possible.append(f'https://www.fifaindex.com{joueur_json[0]["url"]}')
        else:
            lien_absent.append(j)
    return lien_absent, lien_possible
```
